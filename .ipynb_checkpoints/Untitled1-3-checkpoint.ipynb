```python
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import cross_val_score

A = Combined_Values_Recent.values
y = A[:, 5]
y = y.reshape(-1,1)
#print(type(y))     #we want numpy.ndaray
#print(y.shape)     #we want dimensions 1000x1
x = A[:,10]
x = x.reshape(-1,1)

reg = linear_model.LinearRegression()
reg.fit(x,y)
prediction_space = np.linspace(min(x),max(x)).reshape(-1, 1)
plt.scatter(x, y, color='blue')
plt.plot(prediction_space, reg.predict(prediction_space),color='black', linewidth=3)
plt.show()
print("3 Month Libor OLS regression results\n\n" + "No Split R^2 " + str(reg.score(x, y)))
cv_results = cross_val_score(reg, x, y, cv=5, scoring='r2')
print("R^2 of individual splits" + str(cv_results))
print("Mean of splits R^2 " + str(np.mean(cv_results)))

```

```python
from sklearn.model_selection import TimeSeriesSplit
from matplotlib import pyplot

A = Combined_Values.values
y = A[:, 5]
y = y.reshape(-1,1)
x = A[:,8]
x = x.reshape(-1,1)#UNI-----1)
max_size = int(np.size(x,0)/10)
tscv = TimeSeriesSplit(max_train_size = max_size, n_splits = 8)
reg = linear_model.LinearRegression()
index = 1

print("GDP OLS Results using rolling foward cross validation\n")
for train_index, test_index in tscv.split(x):
    
    X_train, X_test = x[train_index], x[test_index]
    y_train, y_test = y[train_index], y[test_index]
    reg.fit(X_train,y_train)
    print("split#" + str(index) + "  self-score " + str(reg.score(X_test, y_test)) + " test-score " + str(reg.score(X_train, y_train)))
    pyplot.subplot(910 + index)
    pyplot.plot(X_train, y_train)
    pyplot.plot(X_test, y_test)
    index += 1
    
print("\nVisualization of training and test data sets")    
pyplot.show()
```
